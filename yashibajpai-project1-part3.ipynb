### Question 1: Compare the race/ethnicity demographics of the survey over time (2020-2024). What are the changes over time, if any? 

```python
""" creates pie chart of the race/ethnicity demographics in the 2024 survey using the matplotlib.pyplot library """

race_counts = fall_2024['race_ethnicity'].value_counts() # counts the occurrences of each unique value in the 'race_ethnicity' column

plt.figure(figsize=(8, 8)) # creates a figure for the pie chart and set the size of the figure (8x8 inches)

# creates the pie chart:
# - 'race_counts' provides the data (frequency of each race/ethnicity)
# - labels=None removes labels from slices to avoid overlap
# - 'autopct' displays percentages on the slices (formatted to 1 decimal place)
# - 'startangle=90' starts the pie chart from a vertical position
plt.pie(race_counts, labels=None, autopct='%1.1f%%', startangle=90) 

# adds a legend outside of the pie chart using race/ethnicity names (index of race_counts):
# - 'loc="best"' automatically finds the best location for the legend
# - 'bbox_to_anchor=(1, 0.5, 0.5, 0.5)' positions the legend outside the pie chart
plt.legend(race_counts.index, loc="best", bbox_to_anchor=(1, 0.5, 0.5, 0.5)) # thank you w3schools for this one

plt.title('Race/Ethnicity Distribution for Fall 2024') # sets the title of the pie chart

plt.axis('equal') # ensures that the pie chart is drawn as a circle by setting the axis equal

plt.show() # displays the pie chart
```

### Question 2: Does a specific gender correlate to higher numbers in the Computer Science and Information Technology degree programs? -

```python
# filters the dataframes for each year to only include rows where 'enrolled_degree_program' contains either 'Computer Science' or 'Information Technology'
filter_2024 = fall_2024[fall_2024['enrolled_degree_program'].isin(['Computer Science', 'Information Technology'])] 
filter_2023 = fall_2023[fall_2023['enrolled_degree_program'].isin(['Computer Science', 'Information Technology'])]
filter_2022 = fall_2022[fall_2022['enrolled_degree_program'].isin(['Computer Science', 'Information Technology'])]
filter_2021 = fall_2021[fall_2021['enrolled_degree_program'].isin(['Computer Science', 'Information Technology'])]
filter_2020 = fall_2020[fall_2020['enrolled_degree_program'].isin(['Computer Science', 'Information Technology'])] 
```

```python
""" creates bar graph showing the counts of each gender in the 'gender' column for all 5 years by inserting the values into a separate data frame """

# counts the gender occurrences in each dataframe
gender_2024 = filter_2024['gender'].value_counts()
gender_2023 = filter_2023['gender'].value_counts()
gender_2022 = filter_2022['gender'].value_counts()
gender_2021 = filter_2021['gender'].value_counts()
gender_2020 = filter_2020['gender'].value_counts()

# creates a new dataframe to hold the counts from each year
gender_counts = pd.DataFrame({
    '2024': gender_2024,
    '2023': gender_2023,
    '2022': gender_2022,
    '2021': gender_2021,
    '2020': gender_2020
})

# plots the bar graph
gender_counts.plot(kind='bar', figsize=(10, 6))

# adds title and labels
plt.title('Gender Distribution from 2020 to 2024')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0) # keeps the x-axis labels horizontal for readability
plt.legend(title='Year') # adds a legend with a title

# shows the plot
plt.show()
```

### • Question 2 Conclusion: 
### It is evident that men consistently make up the largest portion of students in the Computer Science and Information Technology degree programs. Women follow in second place, but their representation is significantly lower.

### Question 3: Does a student taking a middle or high school computing class that had some or a high impact on the decision to enroll in a computing course correlate with a higher level of interest in taking future computing classes?

```python
# filters the dataframes for each year to only include rows where 'impact_hs_computing_class' contains either 'Some Impact' or 'High Impact' 
filter_2024 = fall_2024[fall_2024['impact_hs_computing_class'].isin(['Some Impact', 'High Impact'])] 
filter_2023 = fall_2023[fall_2023['impact_hs_computing_class'].isin(['Some Impact', 'High Impact'])] 
filter_2022 = fall_2022[fall_2022['impact_hs_computing_class'].isin(['Some Impact', 'High Impact'])] 
filter_2021 = fall_2021[fall_2021['impact_hs_computing_class'].isin(['Some Impact', 'High Impact'])] 
filter_2020 = fall_2020[fall_2020['impact_hs_computing_class'].isin(['Some Impact', 'High Impact'])] 
```

```python
# filters the dataframe to exclude rows where 'impact_hs_computing_class' contains 'Not Answered', the tilde negates the condition and selects rows were 'Not Answered' is not found
filter_2024 = filter_2024[~filter_2024['enrolled_course'].str.contains('Not Answered', case=False, na=False)]
filter_2023 = filter_2023[~filter_2023['enrolled_course'].str.contains('Not Answered', case=False, na=False)]
filter_2022 = filter_2022[~filter_2022['enrolled_course'].str.contains('Not Answered', case=False, na=False)]
filter_2021 = filter_2021[~filter_2021['enrolled_course'].str.contains('Not Answered', case=False, na=False)]
filter_2020 = filter_2020[~filter_2020['enrolled_course'].str.contains('Not Answered', case=False, na=False)]
```

```python
# filters the dataframe to exclude rows where 'interest_in_more_classes' contains 'Not Answered', the tilde negates the condition and selects rows were 'Not Answered' is not found
filter_2024 = filter_2024[~filter_2024['interest_in_more_classes'].str.contains('Not Answered', case=False, na=False)]
filter_2023 = filter_2023[~filter_2023['interest_in_more_classes'].str.contains('Not Answered', case=False, na=False)]
filter_2022 = filter_2022[~filter_2022['interest_in_more_classes'].str.contains('Not Answered', case=False, na=False)]
filter_2021 = filter_2021[~filter_2021['interest_in_more_classes'].str.contains('Not Answered', case=False, na=False)]
filter_2020 = filter_2020[~filter_2020['interest_in_more_classes'].str.contains('Not Answered', case=False, na=False)]
```

```python
# concatenates the dataframes into one to make analysis easier
combined_df = pd.concat([filter_2024, filter_2023, filter_2022, filter_2021, filter_2020])

# groups the combined dataframes by 'interest_in_more_classes' and 'enrolled_course'
# - the .size() function counts the occurrences for each combination of 'interest_in_more_classes' and 'enrolled_course'
# - .unstack() makes the 'enrolled_course' values columns, with 'interest_in_more_classes' as the index
# - fill_value=0 ensures that any missing values are filled with 0
group_data = combined_df.groupby(['interest_in_more_classes', 'impact_hs_computing_class']).size().unstack(fill_value=0)

# plots the bar graph
group_data.plot(kind='bar', figsize=(10, 6))

# adds title and labels
plt.title('Count of Interest Ratings by Impact')
plt.xlabel('Interest Rating')
plt.ylabel('Count')
plt.xticks(rotation=0) # keeps the x-axis labels horizontal for readability
plt.legend(title='Impact Level', bbox_to_anchor=(1.05, 1), loc='upper left') # adds a legend with a title

# shows the plot
plt.show()
```

### • Question 3 Conclusion: 
### Students who reported that a middle or high school computing class had "Some Impact" or "High Impact" on their decision to enroll in a computing course show higher interest in future computing classes. Students with "Some Impact" are the most frequent in the 3.0-5.0 interest rating categories, while students with "High Impact" appear more evenly distributed, but still show higher counts at 3.0-5.0 ratings. There are fewer students with low-interest ratings (1.0 and 2.0) in either impact category, suggesting that a positive influence from prior computing experiences correlates with a greater potential interest in future computing classes.
